````python
import random
import math
import matplotlib.pyplot as plt
````

````python

def initializePoints(count):
     points = []
     for i in range(int(count/3)):
         points.append([random.gauss(0,10),random.gauss(100,10)])
     for i in range(int(count/3)):
         points.append([random.gauss(-30,20),random.gauss(10,10)])
     for i in range(int(count/3)):
         points.append([random.gauss(30,20),random.gauss(10,10)])

     return points
````

````python
def EuclideanDistance(p1, p2): 
    dis = math.sqrt(math.pow(p1[0]-p2[0], 2)+math.pow(p1[1]-p2[1], 2))
    return dis
````

````python
def cluster(points,K,visuals = True):
    #Your kmeans code will go here to cluster given points in K clsuters. If visuals = True, the code will also plot graphs to show the current state of clustering
  x = []
  y = []
  for i in points:
    x.append(i[0])
    y.append(i[1])
  centroid = []
  iteration = 1
  clusters = dict()
  for i in range(K):
      index = (random.randint(0,len(x)-1))
      centroid.append((x[index], y[index]))
      clusters[(x[index], y[index])] = []
  check = True
  euclidean_dis = dict()
  for i in range(len(x)):
    euclidean_dis[(x[i], y[i])] = dict()
  while (check == True):
      for i in range(len(x)):
        point = (x[i],y[i])
        for j in range(K):
          eDis = EuclideanDistance(point, centroid[j])
          euclidean_dis[point][centroid[j]] = eDis
      for key2 in euclidean_dis:
        nearby_center = min(euclidean_dis[key2].keys(), key=(lambda k: euclidean_dis[key2][k]))
        clusters[nearby_center].append(key2)
      d = 0
      if visuals == True:
        plt.clf()
        colors = ['red', 'blue', 'green']
        for key in clusters:
          for values in clusters.get(key):
            plt.scatter(values[0], values[1],color = colors[d], marker='+')
          plt.scatter(key[0], key[1], marker = 'o', color = 'black')
          d+=1
        plt.suptitle('Iteration '+str(iteration), fontsize=20)
        plt.xlabel('x', fontsize=18)
        plt.ylabel('y', fontsize=16)
        plt.show()
      print("ITERATION " + str(iteration) + " completed!")
      iteration += 1
      cluster_copy = dict()
      centroid_copy = []
      for key3 in clusters:
        sumx = 0
        sumy = 0
        for i in clusters.get(key3):
            sumx+=i[0]
            sumy+=i[1]
        avg_x = sumx/len(clusters.get(key3))
        avg_y = sumy/len(clusters.get(key3))
        cluster_copy[(avg_x, avg_y)] = []
        centroid_copy.append((avg_x, avg_y))
      if ((centroid[0] == centroid_copy[0]) and (centroid[1] == centroid_copy[1]) and (centroid[2] == centroid_copy[2])):
        check = False
        break
      clusters = dict()
      centroid = []
      for i in centroid_copy:
        centroid.append(i)
        clusters[i] = []
      for key in euclidean_dis:
        euclidean_dis[key] = dict()
  return clusters
````

````python
def clusterQuality(clusters):
    score = 0
    mean_dic = dict()
    #Your code to compute the quality of cluster will go here.
    print("Clusters: ",clusters)
    for key3 in clusters:
        sumx = 0
        sumy = 0
        for i in clusters.get(key3):
            sumx+=i[0]
            sumy+=i[1]
        avg_x = sumx/len(clusters.get(key3))
        avg_y = sumy/len(clusters.get(key3))
        mean_dic[key3] = (avg_x, avg_y)
    for key4 in mean_dic:
      for val in clusters.get(key4):
        ed = EuclideanDistance(val, mean_dic.get(key4))
        score+=pow(ed, 2)
    return score
````

````python
def keepClustering(points,K,N,visuals):
    cluster_score = dict()
    #Write you code to run clustering N times and return the formation having the best quality
    for i in range(N):
      print("N = ", str(i))
      c = cluster(points,K,visuals = visuals)
      q = clusterQuality(c)
      cluster_score[q] = c
    clusters = cluster_score[min(cluster_score)]
    return clusters
    
````

````python
K = 3
N = 10
points = initializePoints(1000)

clusters = keepClustering(points,K,N,True)

print ("The score of best Kmeans clustering is:", clusterQuality(clusters))
````
